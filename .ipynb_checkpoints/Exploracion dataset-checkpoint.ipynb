```python
oportunidad_grouped = df.groupby(['Opportunity_ID', 'ASP_(converted)','Region', 'Stage']).agg({'Total_Amount':['mean', 'count', 'sum']}).reset_index()
```

```python
oportunidad_grouped2 = df.groupby(['Opportunity_ID', 'ASP_(converted)','Region', 'Stage'])['Total_Amount'].sum().reset_index()
```

```python
oportunidad_grouped2['total'] = oportunidad_grouped2['ASP_(converted)'] * oportunidad_grouped2['Total_Amount']
```

```python
oportunidad_grouped2
```

```python
stage = ["Closed Lost", "Closed Won"]

sns.boxenplot(x="Stage", y="Total_Amount",
              color="b", order=stage,
              scale="linear", data=oportunidad_grouped2)
```

```python
region_grouped = df.groupby(['Opportunity_ID','Region', 'Stage'])['Opportunity_Name'].count().reset_index()
```

```python
region_grouped
```

```python
region_grouped.columns = ['Opportunity_ID', 'Region', 'Stage', 'cantidad']
```

```python
region_grouped = region_grouped.groupby(['Region', 'Stage'])['cantidad'].sum().reset_index()
```

```python
region_grouped
```

```python
stage = ["EMEA", "Americas", "Japan", "Middle East"]

sns.boxenplot(x="Region", y="cantidad",
              color="b", order=stage,
              scale="linear", data=region_grouped)
```
